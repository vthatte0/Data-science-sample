```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#Reads data into a dataframe object
df = pd.read_csv('../Data Science Evaluation.csv')

#Adds a column for profit per unit
df['Profit per Unit'] = round(df['Total Profit'].div(df['Units Sold'], axis=0))

#Adds a column for days taken to ship
difference_list = []
for order, ship in zip(list(df["Order Date"]), list(df["Ship Date"])):
    start_date = datetime.strptime(order, '%m/%d/%y')
    end_date = datetime.strptime(ship, '%m/%d/%y')
    difference = end_date - start_date
    difference_list.append(difference.days)
df["Days to Ship"] = difference_list
df



```

```python
#Helper Funtions
def isPercentProfitConstant():
    products = set(df['Item Type'])
    for product in products:
        filteredDf = df[df['Item Type'] == product]
        if len(set(filteredDf['Profit as % of Cost'])) != 1:
            return False
    return True



def isProfitPerUnitConstant():
    products = set(df['Item Type'])
    for product in products:
        filteredDf = df[df['Item Type'] == product]
        if len(set(filteredDf['Profit per Unit'])) != 1:
            return False
    return True

def isSameAmountSold(product):
    regions = set(df['Region'])
    totalForAll = []
    for region in regions:
        filteredDf = df[df['Item Type'] == product]
        totalSold = 0
        for order in filteredDf['Units Sold']:
            totalSold += order
        totalForAll.append(totalSold)
    return len(set(totalForAll)) == 1

def calculateRegression(filtered_df):
    total_x = []
    total_y = []
    for date in set(list(filtered_df["Fiscal Year"])):
        if date != 2017: #Note- 2017 is an outlier for all products and changes the overall trend significantly
            filtered_by_date = filtered_df[filtered_df["Fiscal Year"] == date]
            total_x.append(date)
            total_y.append(filtered_by_date["Total Profit"].sum())
    slope, intercept, r_value, p_value, std_err = stats.linregress(total_x,total_y)
    return slope
    
       
        
            
    
    
    
```

```python
#Question 1(a): Suggest which products should be dropped from selling in the next year
#moving forward and which products should be sold more.

#Method 3: Using linear regression- compare rates of growth or decline

#Note: The year 2017 is excluded- outlier for all products which makes the trend less accurate

products = list(set(df['Item Type']))
trends_by_product = []
for product in products:
    filtered_df = df[df["Item Type"] == product]
    trends_by_product.append(calculateRegression(filtered_df))
trend_zscores = stats.zscore(trends_by_product)
print(trend_zscores)
    
#Products with a slope that is more than 1 standard deviation above the mean should be sold more. Products with a slope
#that is more than 1 standard deviation below the mean should not be sold.
    
sell_more = [products[i] for i in range(0, len(products),1) if trend_zscores[i] > 1]
#Products to sell more: ['Snacks', 'Cereal']
print(sell_more)

sell_less = [products[i] for i in range(0, len(products),1) if trend_zscores[i] < -1]
print(sell_less)
#Products to sell less or stop selling: ['Cosmetics']





```

```python
#Question 1(b): Should any region be given preference over the other?

#Method 1: Consider total profit from each region for each product
    #Note: This strategy assumes that the same amount of each product was sold in each region
    #Note: The assumption that profit per unit sold is constant accross regions has been proven above
    
#Step 1: Confirm the above assumption
products = list(set(df['Item Type']))
for product in products:
    print(isSameAmountSold(product)) #All are true
    
#Step 2: For each product, calculate the total profit by region
regions = list(set(df['Region']))
product_sums = []
for product in products:
    region_sums = []
    product_df = df[df['Item Type'] == product]
    for region in regions:
        region_df = product_df[product_df['Region'] == region]
        region_sum = sum(list(region_df["Total Profit"]))
        region_sums.append(region_sum)
    product_sums.append(region_sums)

#Step 3: Find regions which generate the maximum profit 

prioritizations = []
for i in range(0, len(product_sums),1):
    max_region = regions[product_sums[i].index(max(product_sums[i]))]
    prioritizations.append((products[i], max_region))
print(prioritizations)

#The following list(prioritizations) shows the regions where the selling of each product should be prioritized
#[('Beverages', 'Europe'), ('Vegetables', 'Europe'), ('Cosmetics', 'Sub-Saharan Africa'), 
#('Baby Food', 'Europe'), ('Office Supplies', 'Europe'), ('Personal Care', 'Europe'),
#('Snacks', 'Sub-Saharan Africa'), ('Fruits', 'Europe'), ('Household', 'Europe'), 
#('Cereal', 'Europe'), ('Clothes', 'Sub-Saharan Africa'), ('Meat', 'Europe')]

```

```python
#Question 1(b): Should any region be given preference over the other?

#Method 2- Use linear regression to look at rates of growth and decline in each region

regions = list(set(df['Region']))

product_slopes = []
for product in products:
    region_slopes = []
    for region in regions:
        filtered_df = df[df["Item Type"] == product]
        filtered_df = filtered_df[filtered_df["Region"] == region]
        region_slopes.append(calculateRegression(filtered_df))
    product_slopes.append(region_slopes)

maximum_regions_of_growth = []
for i in range(0, len(products),1):
    max_region = regions[product_slopes[i].index(max(product_slopes[i]))]
    maximum_regions_of_growth.append((products[i], max_region))
print(maximum_regions_of_growth)

#This approach assumes that regions where the total profits of certain products are growing the fastest should be 
#prioritized where those specific products

#This list of tuples shows which region should be prioritized for each product
#[('Cosmetics', 'Asia'), ('Clothes', 'Sub-Saharan Africa'), ('Vegetables', 'Sub-Saharan Africa'), 
#('Household', 'Central America and the Caribbean'), ('Baby Food', 'Central America and the Caribbean'), 
#('Beverages', 'Europe'), ('Cereal', 'Europe'), ('Meat', 'Europe'), ('Snacks', 'Sub-Saharan Africa'), 
#('Fruits', 'Australia and Oceania'), ('Office Supplies', 'Australia and Oceania'), 
#('Personal Care', 'Middle East and North Africa')]
        
```

```python
#Question 2(a): Which products are the easiest to sell?

#Look at the average time difference between order date and ship date for each product
products = list(set(df['Item Type']))
regions = regions = list(set(df['Region']))

region_averages = []

for region in regions:
    region_df = df[df['Region'] == region]
    product_averages = []
    for product in products:
        product_df = region_df[region_df['Item Type'] == product]
        product_averages.append(product_df["Days to Ship"].mean())
    region_averages.append(product_averages)
    
    
easiest_sold = []
for i in range(0, len(regions),1):
    min_product = products[region_averages[i].index(min(region_averages[i]))]
    easiest_sold.append((regions[i], min_product))
print(easiest_sold)

#The following products are easiest to sell in each region
#[('Europe', 'Snacks'), ('Sub-Saharan Africa', 'Meat'), ('Australia and Oceania', 'Fruits'), 
#('Asia', 'Beverages'), ('Central America and the Caribbean', 'Cosmetics'), 
#('North America', 'Cereal'), ('Middle East and North Africa', 'Office Supplies')]


```
